```python
for i in raw_data.columns[raw_data.dtypes == "object"]:
    raw_data=pd.concat([raw_data,pd.get_dummies(raw_data[i],dtype=np.int8,drop_first=True)],axis=1)
    raw_data=raw_data.drop(columns=[i])
```

```python
sdd=StandardScaler()
x_feature=raw_data.iloc[:,:]
x_feature=x_feature.drop(columns='GPA')
y_target=raw_data.loc[:,'GPA']
x_scaled_feature=sdd.fit_transform(x_feature)
x_scaled_feature=pd.DataFrame(x_scaled_feature)
x_scaled_feature
```

```python
poly = PolynomialFeatures(degree=2,include_bias=False)
poly_feature = pd.DataFrame(poly.fit_transform(x_scaled_feature))
poly_feature
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feature,y_target,test_size=0.3,random_state=42)
```

```python
# Model design
# TODO
poly_regressor=LinearRegression()
poly_regressor.fit(x_train,y_train)

y_hat=poly_regressor.predict(x_test)
```

```python
import zipfile
import os

if not os.path.exists(os.path.join(os.getcwd(), 'Student_GPA.ipynb')):
    %notebook -e Student_GPA.ipynb

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)

file_names = ['Student_GPA.ipynb', 'submission.csv']
compress(file_names)
```
